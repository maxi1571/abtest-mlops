# Exploratory analysis

```python
def plot_hist(df:pd.DataFrame, column:str, color:str)->None:
    sns.displot(data=df, x=column, color=color, height=7, aspect=2)
    plt.title(f'Distribution of {column}', size=20, fontweight='bold')
    plt.xticks(rotation=90)

    plt.show()

def plot_count(df:pd.DataFrame, column:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.countplot(data=df, x=column)
    plt.xticks(rotation=90)
    plt.title(f'Distribution of {column}', size=20, fontweight='bold')
    plt.show()
```

```python
# Correlation matrics
df_new=df[['experiment','date','hour','device_make','platform_os','browser','yes','no']].copy()
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
df_new['experiment'] = ord_enc.fit_transform(df_new[["experiment"]])
df_new['device_make'] = ord_enc.fit_transform(df_new[['device_make']])
corrMatrix = df_new.corr()
f, ax = plt.subplots(figsize=(13, 9))
sns.heatmap(corrMatrix,annot=True,fmt='.3f',annot_kws={"size":12})
plt.show()
```

- The biggest correlation is between Platform_OS and Device_Make which is -0.585.

- Experiment has -0.155 correlation between hours
- Experiment has 0.102 correlation between plateform_os
- Experiment has -0.053 correlation between device_make
- Experiment has 0.023 correlation between saying yes
- Experiment has 0.015 correlation between saying no
